```python
f = open('input.txt', 'rt')
total_score = 0
for tx in f:
    print (tx)
    if tx.strip() == "A X":
        total_score += 4
        print ("A X - rock v. rock - Draw")
    elif tx.strip() == "B X":
        total_score += 1
        print ("B X - paper v. rock - Loss")
    elif tx.strip() == "C X":
        total_score += 7
        print ("C X  - scissors v rock - Win")
    elif tx.strip() == "A Y":
        total_score += 8
        print ("A Y - rock v. paper - Win")
    elif tx.strip() == "B Y":
        total_score += 5
        print ("B Y - paper v. paper - Draw")
    elif tx.strip() == "C Y":
        total_score += 2
        print ("C Y  - scissors v paper - Loss")
    elif tx.strip() == "A Z":
        total_score += 3
        print ("A Z - rock v. scissors - Loss")
    elif tx.strip() == "B Z":
        total_score += 9
        print ("B Z - paper v. scissors - Win")
    elif tx.strip() == "C Z":
        total_score += 6
        print ("C Z - scissors v scissors - Draw")
    else:
        print ("Unrecognized pattern ", tx)

print ("total score: ", total_score)
    


    

```
