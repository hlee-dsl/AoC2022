```python
def contained(x, y):
        x_1, x_2 = x.split('-')
        y_1, y_2 = y.split('-')
        x1 = int(x_1)
        x2 = int(x_2)
        y1 = int(y_1)
        y2 = int(y_2)
        if (x1 <= y1 and x2 >= y2) or (x1 >= y1 and x2 <= y2):
            print(x1, x2, y1, y2, "Yes!")
            return True
```

```python
def overlap(x, y):
        x_1, x_2 = x.split('-')
        y_1, y_2 = y.split('-')
        x1 = int(x_1)
        x2 = int(x_2)
        y1 = int(y_1)
        y2 = int(y_2)
        if (x1 <= y1 and y1 <= x2 ) or (y1<=x1 and x1 <= y2):
            print(x1, x2, y1, y2, "Yes!")
            return True
```

```python
f = open('input.txt', 'rt')
total = 0
for s in f:
    print (s)
    x, y = s.split(',')
    if contained(x,y):
        total += 1

print ("Total contained: ", total)
    
```
