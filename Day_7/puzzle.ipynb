```python
def update_msize(tree, current_index, msize):
    # whenever a file size is found, add the size to the appropriate paths - current and all the parents
    if current_index >= 0 :
            tree[current_index][1] += msize

            pid = tree[current_index][2]  # get the parent index
            while pid >= 0:
                tree[pid][1] += msize
                pid = tree[pid][2]


```

```python
# The same code works for both part - set window size to 4 for Part 1
# and 14 for part 2
f = open('input.txt', 'rt')

#initialize key parameters
myTree = []
parent_index = -1
next_index = 0
current_index = -1

for s in f:
    newline = s.rstrip()  # remove white spaces, especially newline "\n"

    if "$ cd" in newline:
        x, y, cd_dir = newline.split(' ')
        # Switching to a new directory - update path
        
        if ".." in cd_dir:
            # go back to the parent directory - reset the variables
            pid = myTree[current_index][2]
            current_path = myTree[pid][0]
            current_index = pid
            parent_path = myTree[myTree[pid][2]][0]
        else:
            # switching to a child directory;  
            if current_index == -1:
                # this is the root
                parent_path = ''
            else:
                parent_path = current_path + "-"
                parent_index = current_index

            # new path - create a new records for the directory
            current_path = parent_path + cd_dir
            new_entry = [current_path, 0, parent_index]
            myTree.append(new_entry)
            current_index = next_index
            next_index += 1
    elif "$ ls" in newline:
        msize = 0  # do nothing
        # loop through the file sizes now
    elif "dir " in newline:
        # do nothing
        #newline.split('')
        msize = 0   # do nothing
    else:
        new_size, file_name = newline.split(' ')  # get the size of the file
        msize = int(new_size) 
        update_msize(myTree, current_index, msize)  # add the new size to the appropriate paths

#### find the sizes that are no more than 100000 and add them up
total_size = 0
for [path, size, parent] in myTree:
    print(path, size, parent)
    if size <= 100000:
        total_size += size


print (total_size)

 

```
